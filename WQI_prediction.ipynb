#From this we can see the best suited algorithm that can work on this dataset
to predict the WQI of water is MULTIPLE LINEAR REGRESSION

```python
X=water_data.iloc[:,0:-2].values
y=water_data.iloc[:,-2].values
```

```python
#Extracting the values of dataset into X for independent variables(column 0 to 7)
#into y for dependent variable(column 8)
```

```python
from sklearn.model_selection import train_test_split
```

```python
#importing train_test_split class from sklearn model_selection 
#module to split the data
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#Splitting the data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
#importing train_test_split class from sklearn linear_model
#module to train the machine using fit method
```

```python
water_data.columns
```

```python
ans=[[29.5,10.15,6.85,4.3,2.35,86.5,159.5,288]]
pred=reg.predict(ans)
pred
```

```python
#prediciting value of wqi by first passing a 2D array using predict method
```

```python
y_pred_all=reg.predict(X)
y_pred_train=reg.predict(X_train)
y_pred_test=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
#importing metrics to calculate accuracy and eror
```

```python
R2_all=r2_score(y,y_pred_all)
R2_train=r2_score(y_train,y_pred_train)
R2_test=r2_score(y_test,y_pred_test)
mse_test=mean_squared_error(y_test, y_pred_test)
```

```python
num_val={'E':5, 'D':4, 'C':3, 'B':2, 'A':1}
water_data['Potability']=water_data['Potability'].map(num_val)
```

```python
#As random forest does not work upon object datatype, we convert each grade into a respective number


#Reference for Water Quality Grade and its potability
# | Range  |  Grade-->Number  | Potability |
#---------------------------------------------
# | 0-25   |     A----->1     | Excellent  |
# | 25-50  |     B----->2     |    Good    |
# | 50-75  |     C----->3     |    Poor    |
# | 75-100 |     D----->4     | Very Poor  |
# | >100   |     E----->5     | Unfit/Unsuitable |
```

```python
water_data
```

```python
water_data['Potability']
```

```python
from sklearn.model_selection import train_test_split

A=water_data.iloc[:,-2:-1].values
B=water_data.iloc[:,-1].values
#In A we extraxt the values of WQI we need to classify and store potability values in B
#through which machine will learn

A_train,A_test,B_train,B_test = train_test_split(A,B,test_size=0.2,random_state=0)

from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier(n_estimators = 5000, oob_score=True)
rfc.fit(A_train,B_train)
B_pred_test=rfc.predict(A_test)
B_pred_train=rfc.predict(A_train)
B_pred_all=rfc.predict(A)
```

```python
accuracy_test=r2_score(B_pred_test, B_test)
accuracy_train=r2_score(B_pred_train, B_train)
accuracy_all=r2_score(B_pred_all, B)
```
